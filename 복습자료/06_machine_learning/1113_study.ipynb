```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 예시 데이터: 타이타닉 데이터셋
df = sns.load_dataset('titanic')

# 1️ 데이터 기본 구조 확인
print(df.shape)        # (891, 15)
print(df.info())       # 데이터 타입 및 결측치 확인

### Dtype -> category를 쓰는 이유 ###
```

```python
# 2️ 통계 요약
print(df.describe())   # 수치형 컬럼의 평균, 표준편차 등
```

```python
# 3️ 결측치 확인
print(df.isna().sum()) # 각 컬럼별 결측치 개수
```

```python
# 4️ 범주형 컬럼의 분포 확인
print(df['sex'].value_counts())  # 남/녀 비율 확인
```

```python
# 5️ 시각화 예시 및 이변량 분석
sns.countplot(x='class', hue='survived', data=df)
plt.title("생존자 비율 by 클래스")
plt.show()
```

```python
data = df[["survived", "pclass"]]

corr_matrix = data.corr()

print("상관관계 표:")
print(corr_matrix)

# 4. heatmap 그리기
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pclass vs Survived Correlation Heatmap")
plt.show()
```

```python
# 클래스, 성별 생존률 알아보기

titanic_rate = df.groupby(["pclass","sex"])["survived"].mean()

print(titanic_rate)
```

## **주요 전처리 단계 요약**

| 번호 | 전처리 종류                 | 설명                   |
| -- | ---------------------- | -------------------- |
| ①  | 결측치 처리 (Missing Value) | 비어있는 값(NaN)을 채우거나 제거 |
| ②  | 이상치 처리 (Outlier)       | 너무 크거나 작은 값 제거       |
| ③  | 스케일링 (Scaling)         | Feature 크기를 일정하게 조정  |
| ④  | 인코딩 (Encoding)         | 문자형 데이터를 숫자로 변환      |
| ⑤  | 데이터 분할 (Split)         | 학습용/테스트용으로 분리        |

```python
import pandas as pd
import numpy as np
np.random.seed(0)

df = pd.DataFrame(np.random.normal(10, 2, size=(10, 3)), columns=['a', 'b', 'c'])
df.iloc[[0, 4], [0, 1]] = [[np.nan, 800],[300,np.nan]]      # 결측치 생성
df.iloc[[0, 3], [0, 2]] = [[100, 2000],[300,-100]]          # 이상치 생성
df
```

```python
# Boxplot을 이용해 이상치 확인 #

import matplotlib.pyplot as plt

df.boxplot()
plt.title('Boxplot')
plt.show()
```

```python
# 결측치 채우기

df[:].fillna(df[:].std(), inplace=True)

# mean()을 사용하지 않은 이유는 이상치와의 차이가 너무 크기 때문에 std()을 사용함
```

```python
df3 = df.copy()

q1 = df3.quantile(0.25)
q3 = df3.quantile(0.75)
iqr = (q3 - q1) * 1.5

lower = q1 - iqr
upper = q3 + iqr

mask = (df3 < lower) | (df3 > upper)

# 중앙값 계산
medians = df3.median()

# 마스크에서 True인 위치에 중앙값 채우기
df3 = df3.mask(mask, df3.median(), axis=1)  #mask가 True이면 df3.median()으로 바꾼다.

df3
```

```python
import pandas as pd

# LabelEncoder는 1차원 자료구조(iterable)을 받아서 변환.
# 숫자의 크기의 차이가 모델에 영향을 주지 않는 트리 계열 모델(의사결정나무, 랜덤포레스트)에 적용한다.
# 숫자의 크기의 차이가 모델에 영향을 미치는 선형 계열 모델(로지스틱회귀, SVM, 신경망)에는 사용하면 안된다.

items = pd.Series(['TV', '냉장고', '컴퓨터', '컴퓨터', '냉장고', '에어콘',  'TV', '에어콘'])
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

# LabelEncoder의 instance 생성
le = LabelEncoder()

# 학습: 각 고유값들을 어떤 정수로 바꿀지 계산.
le.fit(['TV', '냉장고', '컴퓨터', '에어콘', '공기 청정기', '정수기'])  # 인코딩 대상을 넣어 학습한다.

# 변환: 학습 결과에 맞춰서 값들을 변환
result1 = le.transform(items)
print(result1)
```

```python
# 어떤 값을 어떻게 바꿨는지 조회, 값: 고유값, index: encoding 한 값

print(le.classes_)
type(le.classes_)
```

```python
le2 = LabelEncoder()
result2 = le2.fit_transform(items)
print(le2.classes_)
result2

#fit와 transform의 대상이 같다면 fit_transform을 사용
```

```python
import numpy as np

# 원핫 인코딩은 열 단위로 처리하므로 2차원 형태의 자료구조를 입력한다.
# 숫자의 크기 차이가 모델에 영향을 미치는 선형 계열 모델(로지스틱회귀, SVM, 신경망)에서
# 범주형 데이터 변환시 Label Encoding보다 One Hot Encoding을 사용한다.


items=np.array([['TV'],['냉장고'],['전자렌지'],['컴퓨터'],['선풍기'],['선풍기'],['믹서'],['믹서']])
print(np.shape(items)) # items.shape
items  
```

```python
from sklearn.preprocessing import OneHotEncoder
# 객체 생성
ohe = OneHotEncoder()
# 학습 - 어떻게 바꿀지 학습.
ohe.fit(items)
# 변환
result = ohe.transform(items)
# ohe.fit_transform(items)
```

```python
print(result)
```

```python
result.toarray() # ndarray로 변환
```

```python
ohe2 = OneHotEncoder(sparse_output=False)  # ndarray로 결과를 반환.
result2  = ohe2.fit_transform(items)
result2
```

#### **코드 요약**


| 단계         | 코드                   | 설명                      |
| ---------- | -------------------- | ----------------------- |
| **결측치 처리** | `fillna()`           | 비어 있는 값은 평균으로 대체        |
| **이상치 제거** | 조건 필터링               | 너무 큰 salary 값 제거        |
| **인코딩**    | `LabelEncoder()`     | 문자열 데이터(‘M’, ‘F’) → 숫자화 |
| **스케일링**   | `StandardScaler()`   | 값의 단위 차이 제거 (정규화)       |
| **데이터 분할** | `train_test_split()` | 학습용 75%, 테스트용 25%로 분리   |

</br>

#### **스케일링 방식 비교**

| 방법                      | 특징            | 공식                      | 사용 예        |
| ----------------------- | ------------- | ----------------------- | ----------- |
| **표준화(StandardScaler)** | 평균=0, 표준편차=1  | (x - μ) / σ             | 회귀, SVM     |
| **정규화(MinMaxScaler)**   | 0~1 사이 값으로 변환 | (x - min) / (max - min) | 신경망, KNN    |
| **RobustScaler**        | 이상치 영향 줄임     | 중앙값 기반 변환               | 이상치가 많은 데이터 |

</br>

#### **인코딩 방식 비교**

## 3. 데이터 분할 (Train / Val / Test)

</br>
</br>
</br>

#### **훈련용(Train), Val(검증용), Test(테스트용)으로 나눠서 학습**

#### 왜? Overfitting하는 것을 막기 위함

| 구분                       | 용도            | 설명               |
| ------------------------ | ------------- | ---------------- |
| **Train Set (학습용)**      | 모델이 학습하는 데이터  | 패턴을 익히는 단계       |
| **Validation Set (검증용)** | 모델의 성능을 중간 점검 | 하이퍼파라미터 튜닝 등에 사용 |
| **Test Set (테스트용)**      | 최종 성능 평가      | 완전히 새로운 데이터로 검증  |


```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris

X, y = load_iris(return_X_y=True)  # iris.data와 iris.target값만 추출.  (X, y)

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split


## train_test_split(): 2개 dataset 으로 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, # input
    y, # output
    test_size=0.2, # testset의 비율. default: 0.25
    stratify=y, # 분류 데이터셋에만 적용. (y(target)이 범주형) 원본의 클래스들 비율과 동일한 비율로 나누기.
    random_state=42 # random seed값. 나누기 전에 shuffle(섞기)을 먼저 하는 데 그때 일정하게 섞이게 하기 위해 seed값 지정.
)
X_train.shape, X_test.shape
```

```python
### Train set을 두개로 나눠서 하나는 train, 다른 하나는 validation set으로 사용.
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, 
    test_size=0.2,
    stratify=y_train,   # stratify=output(y값, target, label) 지정.
    random_state=0)

X_train.shape, X_test.shape, X_val.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 모델 생성
## max_depth: DecisionTree의 하이퍼 파라미터 중 하나.
max_depth = 1
# max_depth = 2
# max_depth = 3
# max_depth = 4  
tree = DecisionTreeClassifier(max_depth=max_depth, random_state=0)

# 모델 학습 - trainset
tree.fit(X_train, y_train)

# 모델 검증 - validation set /train set
## 1. 예측(추론)
pred_train = tree.predict(X_train)
pred_val = tree.predict(X_val)

## 2. 평가(검증) ==> 정확도
train_acc = accuracy_score(y_train, pred_train)
val_acc = accuracy_score(y_val, pred_val)
```

```python
print(f"max_depth: {max_depth}")
print("Train accuracy:", train_acc)
print("Validation accuracy:", val_acc)
```

```python
max_depth = 4# validation에서 가장 성능 좋은 하이퍼파라미터 사용.
model = DecisionTreeClassifier(max_depth=max_depth, random_state=0)
model.fit(X_train, y_train)
pred = model.predict(X_test)
test_acc = accuracy_score(y_test, pred)
print("최종 평가결과:", test_acc)

# 왜 최종 평가결과 값이 다른가?
```

#### 핵심정리

| 개념                   | 설명                       |
| -------------------- | ------------------------ |
| **train/test split** | 기본적인 데이터 분할 방식           |
| **validation set**   | 하이퍼파라미터 조정용 데이터          |
| **stratify**         | 데이터 비율 유지(특히 분류 문제에서 중요) |
