## 1. 문제 정의 복습 답지



#### 1. C) 문제 정의

머신러닝의 첫 단계는 “무엇을 예측할지”를 정하는 것</br>

문제를 명확히 알아야 어떤 데이터가 필요하고 어떤 모델을 써야 하는지 결정할 수 있습니다</br>
즉, “문제를 수학적으로 표현하는 과정”이 시작점</br>


#### 2. C) 내일 기온 예측

회귀는 연속적인 값을 예측하는 문제</br>
예: 가격, 점수, 온도, 확률 등</br>
반면 “스팸/정상” 같은 건 분류 문제</br>
기온은 숫자(연속값)이므로 회귀 문제에 해당</br>


#### 3. A) 분류 (Classification)

고객이 탈퇴할지(1) / 유지할지(0) → 두 가지 범주로 나뉘니까 분류 문제</br>
즉, 결과가 YES/NO, True/False면 전부 분류 문제</br>



#### 4. B) 어떤 모델을 쓸지 결정하기 위해

문제 정의가 명확해야 어떤 모델(분류/회귀/군집)을 쓸지 선택 가능</br>
만약 문제를 모호하게 잡으면 이후 단계가 전부 꼬임</br>
문제 정의는 머신러닝의 방향키 역할</br>



#### 5. C) 불필요한 데이터 수집 및 시간 낭비

목표가 불분명하면 엉뚱한 데이터를 모으게 됨</br>
예: “매출 예측”인데 고객 리뷰만 모은다든가</br>
결국 “시간 낭비 + 데이터 낭비 + 모델 실패”로 이어짐.</br>

## 2-1. 데이터 수집 복습 답지


#### 1. 정답: B) 학습 재료를 확보

머신러닝은 “데이터 기반 학습”</br>
이 단계의 목적은 모델이 배울 데이터를 모으는 것</br>
아직 모델링(XGBoost나 SVM 등)을 하지 않고, 데이터를 준비하는 단계</br>
즉, “좋은 재료(데이터)가 없으면, 좋은 요리(모델)는 불가능하다”</br>

</br>


#### 2. 정답: A) 최신 데이터 자동 수집 가능

API는 외부 서버에서 실시간 데이터 요청 가능</br>
뉴스, 날씨, 주가처럼 계속 변하는 데이터에 최적</br>
대신 API Key, 호출 횟수 제한 등이 있다는 단점이 있음</br>


</br>



#### 3. 정답: B) HTML 구조 변경에 취약

웹사이트의 구조(HTML, CSS)가 바뀌면</br>
크롤링 코드(soup.select)가 작동 안 함</br>
즉, 유지보수가 자주 필요한 방법</br>

</br>




#### 4. 정답: C) CSV 파일 불러오기

read_csv()는 CSV 파일을 읽어 데이터프레임(DataFrame) 으로 변환하는 함수</br>
머신러닝 입문 시 가장 많이 사용하는 데이터 로딩 방식</br>

</br>




#### 5. 정답: B) 모델 예측력 저하

결측치, 중복, 이상치가 많으면 모델이 잘못된 패턴을 학습해버림.</br>

## 2-2. ERD 복습 답지

</br>

#### 1. 정답: B) 데이터 구조와 특성 파악

EDA는 모델을 만들기 전에 데이터를 “이해”하는 과정</br>
결측치, 이상치, 분포, 상관관계를 파악해서</br>
“이 데이터를 믿고 쓸 수 있는지” 확인하는 게 핵심</br>
즉, 데이터의 건강검진 단계</br>

</br>

#### 2. 정답: B) df.isna().sum()

isna() → 결측 여부(True/False)</br>
.sum() → 각 컬럼별 결측치 개수 계산</br>
df.describe()는 통계 요약용이라 결측치는 안 보여줌</br>
</br>


#### 3. 정답: B) boxplot()

boxplot(상자그림)은 데이터 분포를 시각화하면서</br>
평균에서 너무 멀리 벗어난 “이상치(outlier)”를 한눈에 보줌</br>
점(●)으로 표시된 값이 바로 이상치</br>

</br>

#### 4. 정답: B) df.corr()

corr()은 변수 간의 선형 관계(상관계수)를 계산해줌</br>
예: “공부시간 ↔ 점수” 관계가 강하면 +1에 가까움</br>
heatmap으로 시각화하면 어떤 특징이 서로 영향을 주는지도 볼 수 있음</br>

</br>

#### 5. 정답: B) 모델 학습

모델 학습은 EDA가 끝난 후 “전처리 및 학습 단계”에서 수행함</br>
EDA는 데이터를 “관찰하고 이해하는 단계”일 뿐,</br>
예측은 하지 않음</br>

</br>

## 2-3. 데이터 전처리 복습 답지



#### 1 정답: A) fillna()

fillna()는 비어 있는 데이터를 평균, 중앙값, 특정 값 등으로 채워주는 함수</br>
예시:</br>

>df['age'].fillna(df['age'].mean(), inplace=True)


→ 결측치(NaN)가 있으면 평균값으로 채움.</br>

replace()는 문자열 치환용, concat()은 합치기용이라 전처리와 다름.</br>


</br>


#### 2 정답: B) 상자그림(Boxplot)으로 탐지 후 제거

이상치는 데이터의 평균에서 너무 벗어난 값</br>

boxplot()으로 시각화하면 ‘점’으로 표시돼서 바로 확인 가능</br>

>sns.boxplot(df['salary'])


→ 눈에 띄게 튀는 값이 있으면 제거 or 변환(clip()).</br>

</br>

#### 3 정답: B) Label Encoding

순서형 데이터(예: 초급 < 중급 < 고급)는</br>
각 단계가 순서를 가지므로 숫자로 인코딩하는 게 적합</br>

>encoder = LabelEncoder()

>df['level'] = encoder.fit_transform(df['level'])


초급(0), 중급(1), 고급(2)</br>

반대로 순서가 없는 “성별, 지역” 같은 건 One-Hot Encoding을 사용해야 함</br>


</br>


####  정답: B) 평균 0, 표준편차 1로 변환

StandardScaler()는 데이터의 단위 차이를 없애주는 스케일링 방법</br>

>from sklearn.preprocessing import StandardScaler

>scaler = StandardScaler()

>X_scaled = scaler.fit_transform(X)


평균 = 0, 표준편차 = 1로 맞춰줌.</br>

예: 키(160~180)와 나이(20~60)의 단위 차이를 없애줌.</br>


</br>



####  정답: B) 데이터를 모델이 이해할 수 있도록 정제

머신러닝은 숫자만 이해하니까,</br>
전처리의 목적은 “깨끗한 수치형 데이터로 바꾸는 것”</br>

전처리 없이는 모델이 엉뚱한 방향으로 학습하게 됌</br>

## 3. 데이터 분할 복습 답지

</br>
</br>


####  B) 데이터의 클래스 비율을 유지한다

분류 문제에서 y값 비율이 “불균형”할 수 있음</br>
(예: 정상 90%, 불량 10%).</br>
stratify=y 를 쓰면</br>
→ train/test 나눌 때 클래스 비율을 똑같이 유지</br>
이게 없으면 test 데이터가 너무 치우칠 위험이 있음</br>

</br>



####  C) 모델의 일반화 성능을 안정적으로 평가하기 위해

train/test 한 번만 나누면</br>
분할 운이 안 좋을 수도 있음(편향).</br>
K-Fold는 여러 번 나눠서 평균 성능을 계산하니까</br>
훨씬 믿을 수 있는 평가가 됌</br>
즉, 운빨 제거.</br>


</br>


####  C) 80%

test가 20%면</br>
나머지 80%가 train으로 자동 배정.</br>
그냥 전체의 1에서 test 비율을 뺀 값이 train 비율</br>




</br>

####  정답: B) 5번
K-Fold는 K번 학습 + 평가 반복.</br>
5-Fold라면</br>
→ 학습/검증 과정이 5번 반복되는 구조.</br>


</br>

####  Validation Set의 역할

정답: C) 하이퍼파라미터 조정용</br>
train: 학습</br>
val(검증): 튜닝(조정)용</br>
test: 최종 평가</br>
validation은 모델을 “중간 점검”하는 데이터</br>
